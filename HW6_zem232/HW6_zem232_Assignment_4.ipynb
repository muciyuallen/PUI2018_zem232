## SPLITTING BY GENDER

### H0: There is no statistical difference in trip duration for bikers that ride in the day vs. the night.
(Day: 6 AM - 5:59 PM ; Night: 6 PM - 5:59 AM)
$$ \alpha=0.05$$


```python
# dropping irrelevant columns
df.drop(['stoptime', 'start station id', 'start station name', 'start station latitude',
        'start station longitude', 'end station id', 'end station name',
        'end station latitude', 'end station longitude', 'bikeid', 'usertype', 
        'birth year','gender'], axis=1, inplace=True)
df['tripduration'].dropna(inplace=True)
df.head()
```

```python
df['date'] = pd.to_datetime(df['starttime'])
```

```python
df['night'] = df['date'][(df['date'] >= '00:18:00')&(df['date'] < '00:06:00')]
df['day'] = df['date'][(df['date'] < '00:18:00')&(df['date'] >= '00:06:00')]
df.head()
```

```python
df['hour'] = df['date'].dt.hour 
df['night'] = df['hour'][(df['hour'] >= 18)&(df['hour'] < 6)]
df.head()
```

```python
bins = np.arange(10, 99, 5)
axD = df.day.groupby(pd.cut(df.day, bins)).agg([count_nonzero]).plot(kind='bar', 
                                                                legend=False)
axD.set_title("day riders")

axN = df.night.groupby(pd.cut(df.night, bins)).agg([count_nonzero]).plot(kind='bar',
                                                                legend=False)
axN.set_title("night riders");
```
